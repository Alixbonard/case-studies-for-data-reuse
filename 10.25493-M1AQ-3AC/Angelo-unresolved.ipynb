# Whole cell patch-clamp recordings of cerebellar stellate cells (v1)


In this notebook we will look more closely at the EBRAINS dataset "[Whole cell patch-clamp recordings of cerebellar stellate cells (v1)](https://search.kg.ebrains.eu/?category=Dataset&modality[0]=electrophysiology&dataAccessibility[0]=free%20access&species[0]=Mus%20musculus#3ca4af33-64bd-437a-9c53-2dac19e10168)", contributed by Egidio D’Angelo from department of Brain and Behavioral Sciences, University of Pavia, Pavia, Italy.

As we can see from the dataset description,

<i>In this dataset we investegated whole-cell patch-clamp recordings that were performed in order to reproduce the input/output relationships and to study the specific properties of the cells firing rate.</i>
More information is available in the [Data Descriptor](https://search.kg.ebrains.eu/?category=Dataset&modality[0]=electrophysiology&dataAccessibility[0]=free%20access&species[0]=Mus%20musculus#3ca4af33-64bd-437a-9c53-2dac19e10168).

This dataset forms part of the results reported in Rizza MF, Locatelli F, Masoli S, Sánchez-Ponce D, Muñoz A, Prestori F, D'Angelo E. Stellate cell computational modeling predicts signal filtering in the molecular layer circuit of cerebellum. *Sci Rep.* 2021 Feb **16**;11(1):3873.https://www.nature.com/articles/s41598-021-83209-w

In this notebook we will demonstrate how to access the data files and the metadata from this study, and how these data can be analysed and visualised, with the goal of reproducing some of the figures from the article. 

### Notes: 
Cells used to create certain figures in the publication have specific conditions e.g.: 10 pulses @ 100 Hz-delivered But unable to reproduce the same results with cells under this condition so lacking data or I was unable to exploit the data accurately

```python
#preparing the notebook
import neo
import matplotlib.pyplot as plt
from neo import get_io
from neo import io
import numpy as np
from statistics import mean
import statistics
import glob
```

```python
#presentation of the data
n=0
while n<8:
    plt.rcParams["figure.figsize"] = (7, 5)
    file_path = ("hbp-d000020_PatchClamp-StellateCells_pub/080517_iv.abf")
    reader = get_io(file_path)
    data = reader.read(lazy=True)
    reader = get_io(file_path)
    data = reader.read(lazy=True)
    seg0 = data[0].segments[(n)]
    channel0 = seg0.analogsignals[1]
    channel0 = channel0.load()
    channel0.array_annotations = {}
    plt.plot(channel0.times, channel0, label=f"segment #{n}")
    plt.xlabel(f"Time ({channel0.times.units.dimensionality.string})")
    plt.ylabel(f"Membrane potential ({channel0.units.dimensionality.string})")
    #plt.legend()
    plt.title("")
    plt.show()
    n=n+1
```
