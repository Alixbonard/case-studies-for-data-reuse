```python
import os
import re
import glob
import zipfile
import pickle
import numpy as np
from neo.io import get_io
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from neo import SpikeTrain

import seaborn as sns
import matplotlib.patches as patches
from elephant import spike_train_generation
```

```python
with open('dict_data_CC_step.pkl', 'rb') as fichier:
    dict_data_CC = pickle.load(fichier)
```

```python
seg_count = 0
for file in dict_data_CC.keys():
    signal = dict_data_CC[file]
    for segment in signal[0].segments:
        signal_segmented = segment.analogsignals[0]


        print('file',file) 
        #print('segment:',signal_segmented)
        plt.plot(signal_segmented.times, signal_segmented)
        plt.title('CC step')
        #plt.legend(dict_data_EPSP.keys(),bbox_to_anchor=(1.55, 1.0),loc='upper right')
    plt.show()
```

```python
adapting = "GrC_Subject15_180116/180116_0005 CC step.abf"
non_adapting="GrC_Subject06_090216/090216_0003 CC step.abf"
accelerating="GrC_Subject10_111018/111018_0001 CC step.abf"
listFile=[adapting,non_adapting,accelerating]
```

```python
signal_segmented = {}
time_ms = {}

for file in listFile: 
    data = get_io(file).read(lazy=True)[0] # load data 
    signal_segmented[file] = data.segments[9].analogsignals[0].load().magnitude.flatten() # get the signal in volt of segment 9
    time_ms[file] = data.segments[0].analogsignals[0].load().times * 1000 # convert to ms 
    
```

```python
fig=plt.figure(figsize=(8, 8), dpi=300)
gs = fig.add_gridspec(4, hspace=0)
axs = gs.subplots(sharex=True, sharey=True)

sns.set_theme(style= 'white')
palette = sns.color_palette("deep")
# colors = ['#3f709e', '#5fab99', '#f29744' ] # colors of the article
colors = ['#2b83ba','#1a9641','#fe9929']

legend =["Adapting","Non-Adapting","Accelerating"]

# Voltage 
for i, file in enumerate(listFile):
    voltage = signal_segmented[file]
    time = time_ms[file]
    axs[i].plot(time,voltage, color = colors[i])
    axs[i].spines['top'].set_visible(False)
    axs[i].spines['right'].set_visible(False)
    axs[i].spines['left'].set_visible(False)
    axs[i].spines['bottom'].set_visible(False)
    axs[i].tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
    axs[i].set_ylabel(legend[i])
    axs[i].set_xlim(0,2200)
    axs[i].set_ylim(-80,50)

    r1=axs[i].add_patch(patches.Rectangle((100, -80), 500,130,fill=True, color='gainsboro', edgecolor='gray'))
    r2=axs[i].add_patch(patches.Rectangle((1600,-80),500,130,fill=True, color='gainsboro', edgecolor='darkgray'))


# Simulation 
start_time = 99  # exemple : dbut de la stimulation en ms
end_time = 2100    # exemple : fin de la stimulation en ms

# Ajout d'une stimulation entre start_time et end_time
stimulation_index_start = np.where(time_ms[file] >= start_time)[0][0]
stimulation_index_end = np.where(time_ms[file] <= end_time)[0][-1]

stimulation = np.zeros_like(voltage)
stimulation_active = False  # tat de la stimulation

for i in range(len(voltage)):
    if stimulation_index_start <= i <= stimulation_index_end:
        stimulation[i] = 40  # Taille de la stimulation
plt.plot(time_ms[file], stimulation, label='Stimulation', color='grey')

# Figure 
ax = plt.gca()

# Masquer les pines inutiles
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)


fig.text(0.22, 0.9, r'initial 500ms')
fig.text(0.75, 0.9, r'final 500ms') 
fig.text(0.5, 0.255, r'10pA') 
plt.show()

#todo: legend 20 mV et 500 ms +10 pA 
```
