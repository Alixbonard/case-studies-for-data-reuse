# Studies of data reuse: Dendritic action potentials recordings in human layer 2/3 cortical neurons - derived data (v1)

In this notebook we will look more closely at the EBRAINS dataset "[Dendritic action potentials recordings in human layer 2/3 cortical neurons - derived data (v1)]( https://search.kg.ebrains.eu/?category=Dataset&modality[0]=electrophysiology&dataAccessibility[0]=free%20access&species[0]=Homo%20sapiens#306b1877-a5b3-4663-9e38-f82060020f20)", contributed by Matthew Evan Larkum from the Humboldt University of Berlin, Germany.

As we can see from the dataset description,

<i>This dataset provides the investigation of the dendrites of layer 2 and 3 (L2/3) pyramidal neurons of the human cerebral cortex ex vivo. Dual somato-dendritic patch clamp and two-photon imaging were used to directly investigate the active properties of L2/3 dendrites in acute slices from surgically resected brain tissue of the human neocortex in epilepsy and tumor patients.</i>

More information is available in the [Data Descriptor](https://search.kg.ebrains.eu/instances/306b1877-a5b3-4663-9e38-f82060020f20).

This dataset forms part of the results reported in Gidon A, Zolnik TA, Fidzinski P, Bolduan F, Papoutsi A, Poirazi P, Holtkamp M, Vida I, Larkum ME. Dendritic action potentials and computation in human layer 2/3 cortical neurons. *Science*. 2020 Jan 3;**367(6473)**:83-87.
Masoli, S., Tognolina, M., Laforenza, U. *et al.* (2020) Parameter tuning differentiates granule cell subtypes enriching transmission properties at the cerebellum input stage. *Commun Biol* **3**:222. https://www.science.org/doi/10.1126/science.aax6239?url_ver=Z39.88-2003&rfr_id=ori:rid:crossref.org&rfr_dat=cr_pub%20%200pubmed

In this notebook we will demonstrate how to access the data files and the metadata from this study, and how these data can be analysed and visualised, with the goal of reproducing some of the figures from the article.
