# Studies of data reuse: The contribution of NMDA and AMPA receptors to persistent firing in the dorsolateral prefrontal cortex in working memory

In this notebook we will look more closely at the EBRAINS dataset "[The contribution of NMDA and AMPA receptors to persistent firing in the dorsolateral prefrontal cortex in working memory](https://search.kg.ebrains.eu/?category=Dataset&dataAccessibility[0]=free%20access&species[0]=Macaca%20mulatta&modality[0]=electrophysiology#c57e7440-f403-4f55-8461-a0874c4f45e8)", contributed by Pieter R. Roelfsema from the Netherlands Institute for Neuroscience in Amsterdam, The Netherlands

As we can see from the dataset description,
<i>This dataset provides a comparison the contribution of AMPA-Rs and NMDA-Rs to persistent firing in the41
dorsolateral prefrontal cortex (dlPFC) of male macaque monkeys performing a delayed saccade to a42
memorized spatial location.</i>
More information is available in the [Data Descriptor](https://search.kg.ebrains.eu/instances/c57e7440-f403-4f55-8461-a0874c4f45e8).

This dataset forms part of the results reported in Roelfsema, P. R., Van Vugt, B., Van Kerkoerle, T., & Vartak, D. (2020). The contribution of NMDA and AMPA receptors to persistent firing in the dorsolateral prefrontal cortex in working memory *Journal of Neuroscience* **10** 1523 https://www.jneurosci.org/content/jneuro/early/2020/02/10/JNEUROSCI.2121-19.2020.full.pdf
