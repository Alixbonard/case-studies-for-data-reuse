# Studies of data reuse: Neuronal activity elicited by figure-ground stimuli in primary visual cortex of the awake mouse (v1)

In this notebook we will look more closely at the EBRAINS dataset "[Neuronal activity elicited by figure-ground stimuli in primary visual cortex of the awake mouse (v1)](https://search.kg.ebrains.eu/?category=Dataset&modality[0]=electrophysiology&dataAccessibility[0]=free%20access&species[0]=Mus%20musculus#476f9d47-d7b4-4ee0-a0a1-3ec075a10069)", contributed by Pieter Roelfsema from the Department of Vision and Cognition of Amsterdam, Netherlands.

As we can see from the dataset description,

<i>This dataset of electrophysiological recordings in V1 of awake mice revealed that the responses elicited by figures were stronger than those elicited by the background and even stronger at the edge between figure and background.  </i>
More information is available in the [Data Descriptor](https://search.kg.ebrains.eu/?category=Dataset&modality[0]=electrophysiology&dataAccessibility[0]=free%20access&species[0]=Mus%20musculus#476f9d47-d7b4-4ee0-a0a1-3ec075a10069).

This dataset forms part of the results reported in Schnabel, U.H., Bossens, C., Lorteije, J.A.M. et al. Figure-ground perception in the awake mouse and neuronal activity elicited by figure-ground stimuli in primary visual cortex. *Sci Rep* **8**, 17800 (2018).https://doi.org/10.1038/s41598-018-36087-8

In this notebook we will demonstrate how to access the data files and the metadata from this study, and how these data can be analysed and visualised, with the goal of reproducing some of the figures from the article.
